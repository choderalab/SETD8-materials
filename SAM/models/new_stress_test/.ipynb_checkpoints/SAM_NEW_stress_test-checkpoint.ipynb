```python
import numpy as np
```

```python
def microstates_explored(trajs, time):
    states = []
    for clones_no in list(np.arange(0,len(trajs),50)) + [len(trajs)]:
        states.append([])
        for k in range(5):
            clones_choice = np.random.choice(np.arange(0,len(trajs)), size=clones_no, replace=False)
            dtrajs_ = []
            for i in clones_choice:
                dtrajs_.append(trajs[i][:time])
            if dtrajs_:
                nstates = len(set(np.concatenate(dtrajs_)))
            else:
                nstates = 0
            states[-1].append(nstates)
            
    states_output = []
    for x in states:
        states_output.append(np.mean(x))
        
    return states_output    
```

```python
# last thing - for the heatmaps - get which microstates are in which square exactly
```

```python
def microstates_explored_indexes(trajs, time):
    states = []
    for clones_no in list(np.arange(0,len(trajs),50)) + [len(trajs)]:
        states.append([])
        for k in range(5):
            clones_choice = np.random.choice(np.arange(0,len(trajs)), size=clones_no, replace=False)
            dtrajs_ = []
            for i in clones_choice:
                dtrajs_.append(trajs[i][:time])
            if dtrajs_:
                nstates = list(set(np.concatenate(dtrajs_)))
            else:
                nstates = []
            states[-1].append(nstates)
            
    states_output = []
    for x in states:
        states_output.append(set(np.concatenate(x)))
        
    return states_output    
```

```python
tc_states_indexes = []
sam_states_indexes = []
all_states_indexes = []
```

```python
trajs = tc_trajs
states = tc_states_indexes

for time in np.arange(0,2100,100):
    states_output = microstates_explored_indexes(trajs, time)
    states.append(states_output)
    
np.save('tc_states_indexes', tc_states_indexes)
np.savetxt('tc_states_indexes.csv', tc_states_indexes, delimiter=";", fmt='%s')
```
