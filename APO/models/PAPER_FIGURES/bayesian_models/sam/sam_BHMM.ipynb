```python
its = pyemma.msm.timescales_hmsm(dtrajs, 12, connectivity='largest', errors='bayes')
```

```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
sns.set_context('paper')
sns.set(font='Arial')
sns.set_style('ticks')
```

```python
plt.rcParams['axes.labelweight'] = 'bold'
plt.figure(dpi=300)
ax = pyemma.plots.plot_implied_timescales(its, dt=0.5, units='ns', show_mle=True, show_mean=True, nits=9)
plt.xlabel('lag time/ns')
plt.ylabel('timescale/ns')

# remove the nonconnected data points in the last 4 lines
data = ax.lines[-2].get_data()
data[1][12] = np.nan
ax.lines[-2].set_data(data)

data = ax.lines[-4].get_data()
data[1][12] = np.nan
ax.lines[-4].set_data(data)

data = ax.lines[-6].get_data()
data[1][12] = np.nan
ax.lines[-6].set_data(data)

data = ax.lines[-8].get_data()
data[1][12] = np.nan
ax.lines[-8].set_data(data)

ylim = ax.get_ylim()
plt.vlines(50, ylim[0], ylim[1], colors="red", linestyles="dotted")
ax.set_ylim(ylim)

plt.tight_layout()
plt.savefig('sam_BHMSM_timescales_mean_mle.pdf')
plt.show()
```
