```python
timescales = []

for i in range(23):
    timescales.append([])

errors = []

for i in range(23):
    errors.append([])

for lag in lags:
    try:
        L = np.load('bayes_ck_test_2/%d/timescales_L.npy' % lag)
        R = np.load('bayes_ck_test_2/%d/timescales_R.npy' % lag)
        timescales_ = np.load('bayes_ck_test_2/%d/timescales.npy' % lag)
        for i in range(23):
            try:
                if not (L[i] == np.nan or R[i] == np.nan or timescales_[i] == np.nan):
                    errors[i].append([timescales_[i] - L[i], R[i] - timescales_[i]])
                    timescales[i].append(timescales_[i])
                else:
                    timescales[i].append(np.nan)
                    errors[i].append([np.nan, np.nan])
            except IndexError:
                timescales[i].append(np.nan)
                errors[i].append([np.nan, np.nan])
    except FileNotFoundError:
        for i in range(23):
            timescales[i].append(np.nan)
            errors[i].append([np.nan, np.nan])
```

```python
plt.figure(dpi=300)
for i in range(len(timescales)):
    plt.errorbar(np.arange(100, 2100, 100), timescales[i], yerr=np.array(errors[i]).transpose())
plt.xticks(np.arange(100, 2100, 100))
plt.xlabel('Lag time (steps, 0.5 ns / step)')
plt.ylabel('Timescale (steps, 0.5 ns / step)')
plt.yscale('log')
plt.xticks(rotation='vertical')
plt.show()    
```
