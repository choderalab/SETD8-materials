```python
import sklearn
import sklearn.manifold
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
# let's do a triangle, a square and a circle
```

```python
triangle = np.array([[0,0], [0,5], [4, 2.5]])
```

```python
triangle[:,0]
```

```python
plt.figure(dpi=300)
plt.scatter(triangle[:,0], triangle[:,1])
plt.show()
```

```python
triangle_distances = scipy.spatial.distance.pdist(triangle)
```

```python
np.sum(triangle_distances)
```

```python
triangle_distances = scipy.spatial.distance.squareform(triangle_distances)
```

```python
mds = sklearn.manifold.MDS()
mds_out = mds.fit_transform(triangle)
```

```python
plt.figure(dpi=300)
plt.scatter(mds_out[:,0], mds_out[:,1])
plt.show()
```

```python
mds_distances = scipy.spatial.distance.pdist(mds_out)
```

```python
np.sum(mds_distances)
```

```python
# now try to pass a dissimilarity matrix
```

```python
mds = sklearn.manifold.MDS(dissimilarity='precomputed')
mds_out = mds.fit_transform(triangle_distances)
```

```python
plt.figure(dpi=300)
plt.scatter(mds_out[:,0], mds_out[:,1])
plt.show()
```

```python
mds_distances = scipy.spatial.distance.pdist(mds_out)
```

```python
np.sum(mds_distances)
```

```python
# now try to make mistakes - pass positions with dissimilarity='precomputed' - actually impossible coz matrix wrong shape - or pass distances with dissimilarity = 'euclidean'
```

```python
mds = sklearn.manifold.MDS(dissimilarity='euclidean')
mds_out = mds.fit_transform(triangle_distances)
```

```python
plt.figure(dpi=300)
plt.scatter(mds_out[:,0], mds_out[:,1])
plt.show()
```

```python
mds_distances = scipy.spatial.distance.pdist(mds_out)
```

```python
np.sum(mds_distances)
```

```python
# what if I normalize the dissimilarity (distances) matrix
```

```python
triangle_distances_norm = triangle_distances / np.max(triangle_distances)
```

```python
triangle_distances_norm
```

```python
mds = sklearn.manifold.MDS(dissimilarity='precomputed')
mds_out = mds.fit_transform(triangle_distances_norm)
```

```python
plt.figure(dpi=300)
plt.scatter(mds_out[:,0], mds_out[:,1])
plt.show()
```

```python
mds_distances = scipy.spatial.distance.pdist(mds_out)
```

```python
np.sum(mds_distances)
```

```python
# circle
```

```python
import math
```

```python
def PointsInCircum(r,n=100):
    return [(math.cos(2*np.pi/n*x)*r,math.sin(2*np.pi/n*x)*r) for x in range(0,n+1)]
```

```python
circle = PointsInCircum(5, 10)
circle = np.array(circle)
```

```python
plt.figure(figsize=(4,4), dpi=300)
plt.scatter(circle[:,0], circle[:,1])
plt.show()
```

```python
circle_distances = scipy.spatial.distance.pdist(circle)
```

```python
np.sum(circle_distances)
```

```python
circle_distances = scipy.spatial.distance.squareform(circle_distances)
```

```python
mds = sklearn.manifold.MDS()
mds_out = mds.fit_transform(circle)
```

```python
plt.figure(figsize=(4,4), dpi=300)
plt.scatter(mds_out[:,0], mds_out[:,1])
plt.show()
```

```python
mds_distances = scipy.spatial.distance.pdist(mds_out)
```

```python
np.sum(mds_distances)
```

```python
mds = sklearn.manifold.MDS(dissimilarity='precomputed')
mds_out = mds.fit_transform(circle_distances)
```

```python
plt.figure(figsize=(4,4), dpi=300)
plt.scatter(mds_out[:,0], mds_out[:,1])
plt.show()
```

```python
mds_distances = scipy.spatial.distance.pdist(mds_out)
```

```python
np.sum(mds_distances)
```
