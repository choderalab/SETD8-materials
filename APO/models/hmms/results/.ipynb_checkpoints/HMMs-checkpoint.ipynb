```python
G = nx.Graph()

nodes = []

for i in range(len(metastable_sets_all_hmms)):
    nodes.append([])
    hmm = metastable_sets_all_hmms[i]
    for j in range(len(hmm)):
        nodes[-1].append('%d-%d' % (i+5,j))
    
for node in nodes:
    G.add_nodes_from(node)
 
for a in range(len(nodes)-1):
    nodes1 = nodes[a]
    nodes2 = nodes[a+1]
    meta_sets1 = metastable_sets_all_hmms[a]
    meta_sets2 = metastable_sets_all_hmms[a+1]

    for i in range(len(nodes1)):
        for j in range(len(nodes2)):
            meta_set1 = meta_sets1[i]
            meta_set2 = meta_sets2[j]
            count = 0
            for x in meta_set2:
                if x in meta_set1:
                    count += 1
            weight = count / len(meta_set2)
            if weight > 0.4:
                G.add_edges_from([(nodes1[i], nodes2[j], {'weight': weight})])
```

```python
pos = dict()

for nodes_ in nodes:
    for node in nodes_:
        pos[node] = (-int(node.split('-')[1]), -int(node.split('-')[0]))

fig = plt.figure(figsize=(15,25))
ax = nx.drawing.nx_pylab.draw_networkx(G, pos=pos, node_size=50, with_labels=False)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
for cutoff in np.concatenate([np.arange(0, 0.95, 0.05), [0.99]]):
    G = nx.Graph()

    nodes = []

    for i in range(len(metastable_sets_all_hmms)):
        nodes.append([])
        hmm = metastable_sets_all_hmms[i]
        for j in range(len(hmm)):
            nodes[-1].append('%d-%d' % (i+5,j))
    
    for node in nodes:
        G.add_nodes_from(node)
 
    for a in range(len(nodes)-1):
        nodes1 = nodes[a]
        nodes2 = nodes[a+1]
        meta_sets1 = metastable_sets_all_hmms[a]
        meta_sets2 = metastable_sets_all_hmms[a+1]

        for i in range(len(nodes1)):
            for j in range(len(nodes2)):
                meta_set1 = meta_sets1[i]
                meta_set2 = meta_sets2[j]
                count = 0
                for x in meta_set2:
                    if x in meta_set1:
                        count += 1
                weight = count / len(meta_set2)
                if weight > cutoff:
                    G.add_edges_from([(nodes1[i], nodes2[j], {'weight': weight})])
                
    pos = dict()

    for nodes_ in nodes:
        for node in nodes_:
            pos[node] = (-int(node.split('-')[1]), -int(node.split('-')[0]))

    fig = plt.figure(figsize=(15,25))
    ax = nx.drawing.nx_pylab.draw_networkx(G, pos=pos, node_size=50, with_labels=False)
    plt.xticks([])
    plt.yticks([])
    plt.savefig('macrostate_networks/cutoff_%s.png' % str(cutoff), dpi=300)  
    plt.close()
```

```python
# yep it's a problem
```

```python
np.max([x for x in np.concatenate(inverse_fluxes) if x != np.inf])
```

```python
# actually not only infs are a problem, but anything larger than 1e155 - error, 1e154 - warning, 1e153 - ok. So 
# everything over 1e150 or np.inf I set to 1e150
```

```python
inverse_fluxes_noinf = np.zeros((24,24))

for i in range(24):
    for j in range(24):
        if i == j:
            inverse_fluxes_noinf[i,j] = 0
            continue
        if inverse_fluxes[i,j] == np.inf or inverse_fluxes[i,j] > 1e150:
            inverse_fluxes_noinf[i,j] = 1e150
        else:
            inverse_fluxes_noinf[i,j] = inverse_fluxes[i,j]
```

```python
plt.figure()
plt.imshow(inverse_fluxes_noinf)
plt.colorbar()
plt.show()
```

```python
mds = sklearn.manifold.MDS(dissimilarity='precomputed')
inverse_fluxes_embed = mds.fit_transform(inverse_fluxes_noinf)
```

```python
for j in range(100):
    mds = sklearn.manifold.MDS(dissimilarity='precomputed')
    inverse_fluxes_embed = mds.fit_transform(inverse_fluxes_noinf)
    np.save('inverse_flux_embedding/save_data/%d.npy' % j, inverse_fluxes_embed)
    with open('inverse_flux_embedding/%d.txt' % j, 'w') as f:
        f.write(str(inverse_fluxes_embed))
    f, ax = plt.subplots(dpi=300)
    ax.scatter(inverse_fluxes_embed[:,0], inverse_fluxes_embed[:,1], c='black')
    for i in range(24):
        ax.annotate(i, (inverse_fluxes_embed[i,0], inverse_fluxes_embed[i,1]))
    plt.savefig('inverse_flux_embedding/%d.png' % j, dpi=300)
    plt.close()
```

```python
for j in range(100):
    mds = sklearn.manifold.MDS(dissimilarity='precomputed')
    mfpt_embed = mds.fit_transform(mfpts_sums)
    np.save('mfpt_embedding/save_data/%d.npy' % j, mfpt_embed)
    with open('mfpt_embedding/%d.txt' % j, 'w') as f:
        f.write(str(mfpt_embed))
    f, ax = plt.subplots(dpi=300)
    ax.scatter(mfpt_embed[:,0], mfpt_embed[:,1], c='black')
    for i in range(24):
        ax.annotate(i, (mfpt_embed[i,0], mfpt_embed[i,1]))
    plt.savefig('mfpt_embedding/%d.png' % j, dpi=300)
    plt.close()
```

```python
pi = np.load('/Users/rafalpwiewiora/repos/MSM/11707_11709_SET8_apo_FINAL/dih_full_dset/hmms/results/25/pi.npy')
```

```python
P = np.load('/Users/rafalpwiewiora/repos/MSM/11707_11709_SET8_apo_FINAL/dih_full_dset/hmms/results/25/P.npy')
```

```python
fluxes = P.T * pi
```

```python
P_2 = np.linalg.matrix_power(P, 2)
```

```python
np.count_nonzero(P_2) / (24*24)
```

```python
P_3 = np.linalg.matrix_power(P, 3)
```

```python
np.count_nonzero(P_3) / (24*24)
```

```python
fluxes_3 = np.multiply(P_3.T, pi)
```

```python
inverse_fluxes_3 = 1 / fluxes_3
```

```python
# set diagonals to 0
```

```python
for i in range(24):
    inverse_fluxes_3[i,i] = 0
```

```python
for j in range(100):
    mds = sklearn.manifold.MDS(dissimilarity='precomputed')
    mds_embed = mds.fit_transform(inverse_fluxes_3)
    np.save('inverse_flux_150ns_transmatrix_embedding/save_data/%d.npy' % j, mds_embed)
    with open('inverse_flux_150ns_transmatrix_embedding/%d.txt' % j, 'w') as f:
        f.write(str(mds_embed))
    f, ax = plt.subplots(dpi=300)
    ax.scatter(mds_embed[:,0], mds_embed[:,1], c='black')
    for i in range(24):
        ax.annotate(i, (mds_embed[i,0], mds_embed[i,1]))
    plt.savefig('inverse_flux_150ns_transmatrix_embedding/%d.png' % j, dpi=300)
    plt.close()
```
