```python
def microstates_explored(trajs, time):
    states = []
    for clones_no in list(np.arange(0,len(trajs),50)) + [len(trajs)]:
        states.append([])
        for k in range(5):
            clones_choice = np.random.choice(np.arange(0,len(trajs)), size=clones_no, replace=False)
            dtrajs_ = []
            for i in clones_choice:
                dtrajs_.append(trajs[i][:time])
            if dtrajs_:
                nstates = len(set(np.concatenate(dtrajs_)))
            else:
                nstates = 0
            states[-1].append(nstates)
            
    states_output = []
    for x in states:
        states_output.append(np.mean(x))
        
    return states_output    
```
